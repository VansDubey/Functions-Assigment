```python
"""6. Explain the concept of generators in Python and how they are defined.
Ans=> Generators are functions that yield items one by one using yield, rather than returning all at once. """

#Example:
def gen_numbers():
    for i in range(3):
        yield i

for num in gen_numbers():
    print(num)
```

```python
"""7. What are the advantages of using generators over regular functions?
Ans=> Memory-efficient: Do not store entire sequences in memory.

Lazy evaluation: Values are computed only when needed.

Better for large datasets or infinite sequences. """

#Example:
def infinite_numbers():
    n = 0
    while True:
        yield n
        n += 1
```

```python
"""9. Explain the purpose and usage of the `map()` function in Python.
Ans=> map() applies a function to all items in an iterable and returns a map object (an iterator)."""

#Example:
nums = [1, 2, 3]
squares = list(map(lambda x: x ** 2, nums))
print(squares)  # Output: [1, 4, 9]
```

```python
"""10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Ans=>
map()	applies a function to each item and returns a	map object (iterator)
filter() applies filters items based on a condition and returns a	filter object
reduce() applies function cumulatively to reduce to a single value and result the final result"""
#Example:
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x * 2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
print(reduce(lambda x, y: x + y, nums))
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
```

```python
class Fib:
    def __init__(self, terms):
        self.terms = terms
        self.x = 0
        self.y = 1
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.terms:
            raise StopIteration
        val = self.x
        self.x, self.y = self.y, self.x + self.y
        self.i += 1
        return val

f = Fib(6)
for n in f:
    print(n)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent
```

```python
def pow2(n):
  for i in range(1,n+1):
    yield 2 ** i

for i in pow2(5):
  print(i)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
```

```python
def read_file(f):
  with open(f,'r') as file:
    for line in file:
      yield line.strip()

for i in read_file('/content/sample_data/f.txt'):
  print(i)
```
